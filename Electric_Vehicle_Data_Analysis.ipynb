# **Task 1**
A customer has a budget of 350,000 PLN and wants an EV with a minimum range
of 400 km.

```python
### a) Your task is to filter out EVs that meet these criteria.

EV_filtered =EV[(EV['Minimal price (gross) [PLN]']<=350000) & (EV['Range (WLTP) [km]']>=400)]
EV_filtered
```

```python
### b) Group them by the manufacturer (Make).

for make, group in EV_filtered.groupby("Make"):
  print(f"Manufacturer: {make}")
  print(group)

# groupby functions returns a group oject having tuples with two index which is iterable.
# the first index is the group key which is 'Make' in this case.
# the second index is the all data points in the dataframe corresponding to the key.
```

```python
### c) Calculate the average battery capacity for each manufacturer.

EV_filtered.groupby("Make")["Battery capacity [kWh]"].mean()

# Using groupby(), grouped on the basis of manufacturer
# Then used aggregate function mean() to show battery capacity average for each manufacturer
```

# **Task 2**
 You suspect some EVs have unusually high or low energy consumption.

```python
# Find the outliers in the mean - Energy consumption [kWh/100 km] column.

# Firstly we visualize the data using a boxplot that will also show us if there is a presence of outliers.
sns.boxplot(y=EV['mean - Energy consumption [kWh/100 km]'])
plt.title(f"Boxplot of {'mean - Energy consumption [kWh/100 km]'}")
plt.show()
# We can see there is no such outliers present in our data.
```

```python

# We can say that our data is normally distributed because there is absence of outliers.

mean_col= EV['mean - Energy consumption [kWh/100 km]'].mean()
std_col = EV['mean - Energy consumption [kWh/100 km]'].std()

upr_limit= mean_col+3*std_col
lwr_limit= mean_col-3*std_col
outlier= EV[(EV['mean - Energy consumption [kWh/100 km]']>upr_limit) | (EV['mean - Energy consumption [kWh/100 km]']<lwr_limit) ]
outlier

# 99.7% of the data comes between mean +3*std and mean-3*std of a normal distribution.
# The remaining data is considered outliers.
# In this case there are no such outliers present in the data that comes outside that area.
```

# **Task 4**
**Build an EV recommendation class.**
The class should allow users to input their
budget, desired range, and battery capacity. The class should then return the top three EVs
matching their criteria.


```python
class recommend_EV :
  def __init__(self, data):
    self.data = data
    print("For a recommendation of Top 3 EVs that meet your criteria, call the Top3_EVs_forU function with 3 arguments (budget, desired range, battery capacity)")
# the __init__ method gets called everytime you make a new object.
# Here it suggests you to call a method defined in the class with the arguments you need to enter.

  def Top3_EVs_forU(self, budget, range, battery_capacity):
    personalised_data= self.data[(self.data["Minimal price (gross) [PLN]"]<=budget)&
                                 (self.data["Range (WLTP) [km]"]>=range)&
                                 (self.data["Battery capacity [kWh]"]>=battery_capacity)]
  # personalised_data is the filtered data according to the criteria specified by every object.

    if personalised_data.empty:
      print("No EVs in your criteria.")
      return
    # the fuction terminates with the above string, if the personalised_data is empty.
    else:
      Top3_EVs_forU= personalised_data.sort_values(by=["Range (WLTP) [km]"], ascending= False).head(3)
   # the Top 3 recommendations are sorted from higher to lower on the basis of the range.
   # head function is used to show only the first 3 EVs.
    return Top3_EVs_forU


User1= recommend_EV(EV)
User1.Top3_EVs_forU(375000,350,90)
```

# **Task 5**
Inferential Statistics – Hypothesis Testing: Test whether there is a significant
difference in the average Engine power [KM] of vehicles manufactured by two leading
manufacturers i.e. Tesla and Audi. What insights can you draw from the test results?
Recommendations and Conclusion: Provide actionable insights based on your analysis.
(Conduct a two sample t-test using ttest_ind from scipy.stats module)

```python
from scipy.stats import ttest_ind

tesla_pow = EV[EV["Make"]=="Tesla"]["Engine power [KM]"].dropna()
audi_pow = EV[EV["Make"]=="Audi"]["Engine power [KM]"].dropna()

t_stat, p_val = ttest_ind(tesla_pow, audi_pow)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}\n")

alpha= 0.05
if p_val< alpha:
  print("Reject the null hypothesis, there is a significant difference between average Engine power of Tesla EVs and Audi EVs.")
  print("Recommendation: Further investigation is needed to determine the underlying factors contributing to the difference in engine power between Tesla and Audi EVs.")

else:
  print("Fail to reject the null hypothesis, there is no significant difference between average Engine power of Tesla EVs and Audi EVs.")
  print("Recommendation: No significant difference in power has been observed, it is advisable to direct marketing and product strategy efforts toward other distinguishing attributes.")

```
